```python
import requests
from bs4 import BeautifulSoup
```

```python
page=requests.get("http://top250-movies.surge.sh")
```

```python
soup = BeautifulSoup(page.text, 'html.parser')
```

```python
print(soup)
```

```python
Rates=soup.find_all('span',class_="certificate")
```

```python
print(Rates[0].text)
```

```python
Rates1=[]
for i in range(0,len(Rates)):
    
    Rates1.append(Rates[i].text)
    i=i+1
print(len(Rates1))
```

```python
movie_name=soup.find_all('a',href=True)
#print(movie_name[0].text)
#print(movie_name[1].text)
#print(movie_name[2].text)
```

```python


del movie_name[0]
del movie_name[0]
print(movie_name[0].text)
print(len(movie_name))
```

```python
movie_name1=[]
for i in range(0,int(len( movie_name)/2)):
    movie_name1.append( movie_name[2*i].text)
    i=i+2
print(len(movie_name1))
```

```python
runtime=soup.find_all('span',class_="runtime")
print(runtime[0].text)
```

```python
runtime1=[]
for i in range(0,len(runtime)):
    
    runtime1.append(runtime[i].text)
    i=i+1
print(len(runtime1))   
```

```python
genre=soup.find_all('span',class_="genre")
print(genre[0].text)
print(len(genre))
```

```python
genre1=[]
for i in range(0,len(genre)):
    
    genre1.append(genre[i].text)
    i=i+1
print(len(genre1))
```

```python
#about1=soup.find_all('div',class_="ratings-bar")
about=soup.find_all('p',class_="text-muted")
print(about[1].text)  #1,3,5
print(len(about))
```

```python
about1=[]
for i in range(0,int(len(about)/2)):
    about1.append(about[2*i+1].text)
    i=i+2
print(len(about1))
```

```python
dir=soup.find_all('p',class_="")
print(dir[0].text.split(':')[1].strip())
print(len(dir))
```

```python
dir1=[]
for i in range(0,len(dir)):
    dir1.append(dir[i].text.split(':')[1].strip())
    i=i+1
    
print(len(dir))
```

```python
votes=soup.find_all('p',class_="sort-num_votes-visible")
print(votes[0].text.split(':')[1].strip())
print(len(votes))
```

```python
votes1=[]
for i in range(0,len(votes)):
    votes1.append(votes[i].text.split(':')[1].strip())
    i=i+1
print(len(votes1))
```

```python

import pandas as pd
data1=pd.DataFrame({'movie name':movie_name1,'duration':runtime1,'genre':genre1,'ratings':Rates1,'description':about1,'Director':dir1,'votes':votes1})
```
